```python
# create a Stamen Toner map of the world centered around India
world_map = folium.Map(location=[20.5937, 78.9629], zoom_start=4, tiles='Stamen Toner')

# display map
world_map
```

```python
cases=folium.map.FeatureGroup()
for lat,lng in zip(df_covid.lat,df_covid.lon):
        cases.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
world_map.add_child(cases)
```

```python
# instantiate a feature group for the cases in the dataframe
cases = folium.map.FeatureGroup()

# loop through the cases and add each to the cases feature group
for lat, lng, in zip(df_covid.lat, df_covid.lon):
    cases.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add pop-up text to each marker on the map
latitudes = list(df_covid.lat)
longitudes = list(df_covid.lon)
labels = list(df_covid.confirmed)
labels=str(labels)
for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(world_map)    
    
# add cases to map
world_map.add_child(cases)
```

In the above Map we could visualize the confirmed covid cases(till 1/22/20) by touching the plot.We could Zoom in and out.
